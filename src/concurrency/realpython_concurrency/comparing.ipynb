```python
from typing import Callable
import time


def print_run_time(func: Callable):
    start_time = time.perf_counter()
    func()
    duration = time.perf_counter() - start_time
    print(f"Function {func.__name__} took {duration:.8f} seconds to run")
    return duration
```

```python
from concurrency.realpython_concurrency import (
    cpu_non_concurrent,
    cpu_threads,
    cpu_asyncio,
    cpu_processes,
)
```

```python
print_run_time(cpu_processes.main)
```

```python
print_run_time(cpu_non_concurrent.main)
```

```python
print_run_time(cpu_threads.main)
```

```python
start_time = time.perf_counter()
await cpu_asyncio.main()
duration = time.perf_counter() - start_time
print(f"Function io_asyncio.main took {duration:.8f} seconds to run")
```
